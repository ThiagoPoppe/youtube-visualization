```python
import os

import plotly

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Criando diretório para salvarmos as visualizações
os.makedirs('pequenos_multiplos', exist_ok=True)
```

## Lendo os dados

```python
us_data = pd.read_csv('../../YoutubeProcessedData/USdata.csv', parse_dates=['publish_time', 'trending_date'])
us_data.head()
```

## Definindo espectro de cores

```python
# Gerando um array de cores do arco-íris fixando a saturação e brilho da
# representação HSL da cor e caminhando pela matiz.
category_colors = [f'hsl({h},50%,50%)' for h in np.linspace(0, 360, 30)]
```

## Construíndo a visualização

### 1) Número de vídeos que entraram em trending mais de uma vez

```python
trend_count = us_data.groupby(['title', 'category_name']).size()
trend_count = trend_count.to_frame('trend_count')
trend_count = trend_count.reset_index()

trend_count = trend_count[trend_count['trend_count'] > 1]
trend_count.head()
```

```python
fig, axs = plt.subplots(nrows=4, ncols=4, figsize=(12,10))

categories = sorted(trend_count['category_name'].unique())
categories = np.array(categories).reshape((4,4))

for i in range(4):
    for j in range(4):
        category = categories[i][j]
        axs[i,j].set_title(f'{category}', fontsize=14)
        axs[i,j].hist(trend_count[trend_count['category_name'] == category]['trend_count'])

fig.suptitle('Número de vídeos que entraram em trending mais de uma vez', fontsize=24)

# add a big axis, hide frame
fig.add_subplot(111, frameon=False)
fig.tight_layout()

# hide tick and tick label of the big axis
plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False)
plt.xlabel('Quantidade de dias em trend', fontsize=20)
plt.ylabel('Número de vídeos', fontsize=20)
plt.show()
```

```python
# Salvando a visualização
fig.savefig('pequenos_multiplos/trend_count.png')
```

### 2) Tempo necessário para o vídeo entrar em trend

```python
def get_trend_start_time(df):
    return (df['trending_date'].iloc[0] - df['publish_time'].iloc[0]).days

trend_start_time = us_data.groupby(['title', 'publish_time']).apply(get_trend_start_time)
trend_start_time = trend_start_time.to_frame('trend_start_time')
trend_start_time = trend_start_time.reset_index()
trend_start_time = trend_start_time.drop('publish_time', axis=1)

trend_start_time = trend_start_time.merge(us_data[['title', 'category_name']])
trend_start_time = trend_start_time.drop_duplicates()

trend_start_time = trend_start_time.set_index(['title', 'category_name'])
trend_start_time.head()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12,10))

ax1.set_title('Histograma padrão', fontsize=16)
ax1.set_ylabel('Contagem de vídeos', fontsize=14)
ax1.set_xlabel('Tempo necessário (em dias)', fontsize=14)
ax1.hist(trend_start_time.values)

ax2.set_title('Histograma xlog', fontsize=16)
ax2.set_ylabel('Contagem de vídeos', fontsize=14)
ax2.set_xlabel('Tempo necessário > 0 (escala log10)', fontsize=14)
ax2.hist(np.log10(trend_start_time[trend_start_time > 0].values))

ax3.set_title('Histograma ylog', fontsize=16)
ax3.set_ylabel('Contagem de vídeos (escala log10)', fontsize=14)
ax3.set_xlabel('Tempo necessário (em dias)', fontsize=14)
ax3.hist(trend_start_time, log=True)

ax4.set_title('Histograma log-log)', fontsize=16)
ax4.set_ylabel('Contagem de vídeos (escala log10)', fontsize=14)
ax4.set_xlabel('Tempo necessário > 0 (escala log10)', fontsize=14)
ax4.hist(np.log10(trend_start_time[trend_start_time > 0].values), log=True)

fig.suptitle('Tempo necessário em dias para o vídeo entrar em trend', fontsize=24)
fig.tight_layout()
```

```python
# Salvando a visualização
fig.savefig('pequenos_multiplos/trend_start_time.png')
```

### 3) Maior janela temporal onde um vídeo ficou em trend por categoria

```python
def biggest_window_trend_time(df):
    max_window = 1
    diff1 = df['trending_date'].diff() == pd.Timedelta('1D')
    
    counter = 0
    for value in diff1:
        counter = counter + 1 if value else 1
        max_window = max(max_window, counter)
        
    return max_window

max_trend_window = us_data.groupby(['title', 'category_name']).apply(biggest_window_trend_time)
max_trend_window = max_trend_window.to_frame('max_trend_window')
max_trend_window = max_trend_window.reset_index()

max_trend_window.head()
```

```python
fig, axs = plt.subplots(nrows=4, ncols=4, figsize=(12,10))

categories = sorted(trend_count['category_name'].unique())
categories = np.array(categories).reshape((4,4))

for i in range(4):
    for j in range(4):
        category = categories[i][j]
        axs[i,j].set_title(f'{category}', fontsize=14)
        axs[i,j].hist(max_trend_window[max_trend_window['category_name'] == category]['max_trend_window'])

fig.suptitle('Maior janela de trend para os vídeos de cada categoria', fontsize=24)

# add a big axis, hide frame
fig.add_subplot(111, frameon=False)
fig.tight_layout()

# hide tick and tick label of the big axis
plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False)
plt.xlabel('Tamanho da janela máxima (em dias)', fontsize=20)
plt.ylabel('Número de vídeos', fontsize=20)
plt.show()
```

```python
# Salvando a visualização
fig.savefig('pequenos_multiplos/max_trend_window.png')
```
