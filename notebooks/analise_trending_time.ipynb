```python
from os.path import join as ospj

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Insira aqui o caminho para os dados processados!
BASEPATH = '../YoutubeData/processed/'
```

# Funções auxiliares

```python
def head(df, n=5):
    """ Função auxiliar para exibir as 'n' primeiras linhas dos dados e o seu formato """
    print('Formato dos dados:', df.shape)
    display(df.head(n))
    
def load_data(country):
    """ Função auxiliar para carregar os dados de um país, convertendo as datas no formato datetime """
    filepath = ospj(BASEPATH, f'{country}data.csv')
    return pd.read_csv(filepath, parse_dates=['trending_date', 'publish_time'])
```

```python
us_data = load_data('US')
head(us_data)
```

# 1) Visualização de um vídeo específico ao longo do tempo

## 1.1) Filtrando os dados do vídeo

```python
video_title = 'Childish Gambino - This Is America (Official Video)'

video_data = us_data[us_data['title'] == video_title]
video_data = video_data.sort_values(by='trending_date')

head(video_data)
```

## 1.2) Dias em que o vídeo ficou em trend

```python
publish_time = video_data['publish_time'].values[0]
trending_dates = sorted(video_data['trending_date'].unique())

plt.title(video_title)
plt.scatter(publish_time, y=0, color='blue', label='publish date')
plt.scatter(trending_dates, y=np.zeros(len(trending_dates)), color='red', label='trending date(s)')

plt.legend()
plt.xticks(rotation=45)
plt.show()
```

## 1.3) Evolução das views ao longo do tempo

```python
plt.title('Evolução das views ao longo do tempo')
plt.ylabel('Número de views')
plt.plot(video_data['trending_date'], video_data['views'])
plt.xticks(rotation=45)
plt.show()
```

## 1.4) Evolução de likes e dislikes views ao longo do tempo de trend

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(11,4))

ax1.set_title('Evolução de likes e dislikes ao longo do tempo')
ax1.set_ylabel('Número de likes e dislikes')
ax1.tick_params(labelrotation=45)
ax1.plot(video_data['trending_date'], video_data['likes'], label='likes', color='blue')
ax1.plot(video_data['trending_date'], video_data['dislikes'], label='dislikes', color='red')

ax2.set_title('Taxa de likes e dislikes ao longo do tempo')
ax2.set_ylabel('Log do número de likes e dislikes')
ax2.tick_params(labelrotation=45)
ax2.plot(video_data['trending_date'], np.log(video_data['likes']), label='likes', )
ax2.plot(video_data['trending_date'], np.log(video_data['dislikes']), label='dislikes', color='red')

fig.tight_layout()
fig.show()
```

# 2) Visualizações dos vídeos dos Estados Unidos

## 2.1) Tempo necessário para o vídeo entrar em trend
- Note que vídeos, especialmente de trailers, possuem vários vídeos com o mesmo nome... Por isso, para evitar análises inconsistentes, devemos agrupar os dados em título e data de publicação antes de calcular o tempo necessário.

```python
def get_trend_start_time(df):
    return min(df['trending_date'].values) - df['publish_time'].values[0]

trend_start_time = us_data.groupby(['title', 'publish_time']).apply(get_trend_start_time)
trend_start_time = trend_start_time.astype('timedelta64[D]')

trend_start_time.sort_values()
```

```python
# Esse vídeo demorou 2 dias para entrar em trend
trend_start_time['Childish Gambino - This Is America (Official Video)']
```

```python
# Exemplo de um vídeo com duas datas de publicação diferentes
trend_start_time['Mission: Impossible - Fallout (2018) - Official Trailer - Paramount Pictures']
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,8))

ax1.set_title('Tempo mínimo para entrar em trend (histograma padrão)')
ax1.set_ylabel('Contagem de vídeos')
ax1.set_xlabel('Tempo mínimo (em dias)')
ax1.hist(trend_start_time.values)
  
ax2.set_title('Tempo mínimo para entrar em trend (histograma xlog)')
ax2.set_ylabel('Contagem de vídeos')
ax2.set_xlabel('Tempo mínimo (em dias, log10)')
ax2.hist(np.log10(trend_start_time[trend_start_time > 0].values))

ax3.set_title('Tempo mínimo para entrar em trend (histograma ylog)')
ax3.set_ylabel('Contagem de vídeos (log10)')
ax3.set_xlabel('Tempo mínimo (em dias)')
ax3.hist(trend_start_time, log=True)

ax4.set_title('Tempo mínimo para entrar em trend (histograma log-log)')
ax4.set_ylabel('Contagem de vídeos (log10)')
ax4.set_xlabel('Tempo mínimo (em dias, log10)')
ax4.hist(np.log10(trend_start_time[trend_start_time > 0].values), log=True)

fig.tight_layout()
fig.show()
```

## 2.2) Maior janela temporal onde um vídeo ficou em trend por categoria

```python
def biggest_window_trend_time(df):
    max_window = 1
    diff1 = df['trending_date'].diff() == pd.Timedelta('1D')
    
    counter = 0
    for value in diff1:
        counter = counter + 1 if value else 1
        max_window = max(max_window, counter)
        
    return max_window

trend_windows = us_data.groupby(['title', 'category_name']).apply(biggest_window_trend_time)
trend_windows = trend_windows.reset_index(name='window_size')

head(trend_windows)
```

```python
fig, axs = plt.subplots(nrows=4, ncols=4, figsize=(10,8))

categories = trend_windows['category_name'].unique().reshape((4,4))
for i in range(4):
    for j in range(4):
        category = categories[i][j]
        axs[i,j].set_title(f'{category}')
        axs[i,j].hist(trend_windows[trend_windows['category_name'] == category]['window_size'])

fig.suptitle('Histogramas da maior janela de trend para os vídeos de cada categoria', fontsize=14)
fig.tight_layout()
fig.show()
```

```python
fig, axs = plt.subplots(nrows=4, ncols=4, figsize=(10,8))

categories = trend_windows['category_name'].unique().reshape((4,4))
for i in range(4):
    for j in range(4):
        category = categories[i][j]
        axs[i,j].set_title(f'{category}')
        axs[i,j].boxplot(trend_windows[trend_windows['category_name'] == category]['window_size'])

fig.suptitle('Boxplots da maior janela de trend para os vídeos de cada categoria', fontsize=14)
fig.tight_layout()
fig.show()
```
